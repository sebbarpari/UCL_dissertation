# Features for comparing to baseline movements

```python
def walkingfeatures(spec, person, rep, label, irr):
    main_line=np.zeros((spec.shape[1]))
    pos=np.zeros((spec.shape[1]))
    length=[]
    len_ev=0
    for i in range(spec.shape[1]):
        #LENGTH OF WALK
        down=spec[0:800,i]
        up=spec[801:1600,i]
        if abs(sum(up))>abs(sum(down)):
            pos[i]=1
        else:
            pos[i]=0
        if i>0 and pos[i]==pos[i-1]:
            len_ev=len_ev+1
        else:
            length.append(len_ev)
            len_ev=1

        #MAIN LINE 
        for j in range(spec.shape[0]//2):
            if pos[i]==1 and spec[(spec.shape[0]//2)-j][i]>-16:
                ind=(spec.shape[0]//2)-j
            if pos[i]==0 and spec[(spec.shape[0]//2)+j][i]>-16:
                ind=(spec.shape[0]//2)+j
        main_line[i]=ind

    #Length of movement
    length.append(len_ev)
    start=0
    for i in range(len(length)):
        if length[i]<300:
            start=start+length[i]
        else:
            break
    length=[s for s in length if s>300]
    av_len=np.mean(length)

    main_move=main_line[start:start+sum(length)]
    pos1=pos[start:start+sum(length)]

    maximums=[]
    minimums=[]
    
    #MAXS and MINS
    for i in range(15,len(main_move)):
        line=main_move[i-15:i+16]
        if line[len(line)//2+1]==line.max():
            maximums.append(i)
        if line[len(line)//2+1]==line.min():
            minimums.append(i)
    #Delete repeated values
    delete=[]
    for i in range(len(maximums)-1):
            if maximums[i]-maximums[i+1]>-10:
                delete.append(i)
    for index in sorted(delete, reverse=True):
        del maximums[index]
    delete=[]
    for i in range(len(minimums)-1):
            if minimums[i]-minimums[i+1]>-10:
                delete.append(i)
    for index in sorted(delete, reverse=True):
        del minimums[index]

    #Peak walking and leg velocity
    peak_leg_low=0
    peak_walk_low=0
    peak_walk_high=0
    peak_leg_high=0
    if 1 in pos1 and 0 in pos1:
        peak_leg_low=800-main_move[minimums].min()
        peak_walk_low=800-main_move[maximums].min()
        peak_leg_high=main_move[maximums].max()-800
        peak_walk_high=main_move[minimums].max()-800
        peak_leg=(peak_leg_low+peak_leg_high)/2
        peak_walk=(peak_walk_low+peak_walk_high)/2
    else:
        if 0 in pos1:
            peak_leg=main_move[maximums].max()-800
            peak_walk=main_move[minimums].max()-800
        else: 
            if 1 in pos1:
                peak_leg=800-main_move[minimums].min()
                peak_walk=800-main_move[maximums].min()

    #Time between steps
    dist_min=[]
    for i in range(1,len(minimums)):
        if pos[minimums[i]]==1:
            dist_min.append(minimums[i]-minimums[i-1])
    dist_max=[]
    for i in range(1, len(maximums)):
        if pos[maximums[i]]==0:
            dist_max.append(maximums[i]-maximums[i-1])
    dist=dist_max+dist_min
    dist=np.array(dist)
    #Delete outliers
    mean = np.mean(dist)
    standard_deviation = np.std(dist)
    distance_from_mean = abs(dist - mean)
    max_deviations = 2
    not_outlier = distance_from_mean < max_deviations * standard_deviation
    final_dist = dist[not_outlier]
    dist_steps=np.mean(final_dist)

    #Store in dataframe
    d = {'Person': person, 
        'Repetition': rep,
        'Activity': label,
        'Irregularity' :irr,
        'Dist_steps': dist_steps,
        'Peak_walk': peak_walk,
        'Peak_legs': peak_leg,
        'Len_move': av_len
        }

    try:
        df2=pd.DataFrame(data=d, index=[0])
        df=df.append(df2, ignore_index=True)
    except: 
        df=pd.DataFrame(data=d, index=[0])
        
    return df
```

```python
def sittingfeatures(spec, person, rep, label, irr):
    main_line=np.zeros((spec.shape[1]))
    pos=np.zeros((spec.shape[1]))
    df= pd.DataFrame(columns = ['Person', 'Repetition', 'Activity','Irregularity', 'Max_velocity', 'Dropoff', 'Len_move'])
    length=[]
    len_ev=0
    for i in range(spec.shape[1]):
        #LENGTH OF MOVEMENT
        down=spec[0:800,i]
        up=spec[801:1600,i]
        if abs(sum(up))>abs(sum(down)):
            pos[i]=1
        else:
            pos[i]=0
        if i>0 and pos[i]==pos[i-1]:
            len_ev=len_ev+1
        else:
            length.append(len_ev)
            len_ev=1

        #MAIN LINE 
        for j in range(spec.shape[0]//2):
            if pos[i]==1 and spec[(spec.shape[0]//2)-j][i]>-16:
                ind=(spec.shape[0]//2)-j
            if pos[i]==0 and spec[(spec.shape[0]//2)+j][i]>-16:
                ind=(spec.shape[0]//2)+j
        main_line[i]=ind
        
    if label=='2':
        #Max value
        max_velocity=main_line.max()-800
        #Dropoff
        vel_arg=np.argmax(main_line)
        for i in range(vel_arg, len(main_line)):
            if main_line[i]<=815:
                dropoff_arg=i
                break
    else:
        #Max_value
        max_velocity=800-main_line.min()
        #Dropoff
        vel_arg=np.argmin(main_line)
        for i in range(vel_arg, len(main_line)):
            if main_line[i]>=785:
                dropoff_arg=i
                break
                
    dropoff=dropoff_arg-vel_arg
    
    #length of event
    l=0
    
    for i in range(spec.shape[1]):
        if(abs(np.amax(spec[:,i]))<20):
            l=l+1
    ev_len=l/spec.shape[1]
    
    #Store in dataframe
    d = {'Person': person, 
        'Repetition': rep,
        'Activity': label,
        'Irregularity' :irr,
        'Max_velocity': max_velocity,
        'Dropoff': dropoff,
        'Len_move': ev_len
        } 
        
    
    df=pd.DataFrame(data=d, index=[0])
    return df
```

```python
def getparameters(direct):
    df_walk = pd.DataFrame(columns = ['Person', 'Repetition', 'Activity','Irregularity', 'Dist_steps', 'Peak_walk', 'Peak_legs', 'Len_move'])
    df_sit = pd.DataFrame(columns = ['Person', 'Repetition', 'Activity','Irregularity', 'Max_velocity', 'Dropoff', 'Len_move'])
    df_stand = pd.DataFrame(columns = ['Person', 'Repetition', 'Activity','Irregularity', 'Max_velocity', 'Dropoff', 'Len_move'])
       
    for file in os.listdir(direct):
        spec = np.genfromtxt(direct+r'/'+file, delimiter=',')
        
        #Extract label, irr, person and repetition
        label = file.split('P')[0]
        irr = file.split('I')[1][0]
        rep = int( file.split('R')[1][0])
        if rep==0:
            rep = int( file.split('R')[1][1])
        person = int(file.split('P')[1].split('I')[0])
        
        #Based on the movement, extract different features
        if label=='1':
            try:
                df2=walkingfeatures(spec, person, rep, label, irr)
                df_walk=df_walk.append(df2, ignore_index=True)
            except:
                pass
        else:
            if label=='2':
                try:
                    df2=sittingfeatures(spec, person, rep, label, irr)
                    df_sit=df_sit.append(df2, ignore_index=True)
                except:
                    pass
            else:
                try:
                    df2=sittingfeatures(spec, person, rep, label, irr)
                    df_stand=df_stand.append(df2, ignore_index=True)
                except:
                    pass
    return df_walk, df_sit, df_stand
```

```python
#Obtain mean and std of baseline movements
df_walk, df_sit, df_stand=getparameters('D:/Captured_MTI')

#Walking
d1=df_walk.groupby(by=["Person", "Irregularity"]).mean() #Mean stored in one dataframe
d1=d1.drop(index=['1','2','3','4','5','6','7'], level=1)
d2=df_walk.groupby(by=["Person", "Irregularity"]).std() #Std store in one dataframe
d2=d2.drop(index=['1','2','3','4','5','6','7'], level=1)
d2 = d2.rename(columns={'Dist_steps': 'std_steps', 'Peak_walk': 'std_walk', 'Peak_legs': 'std_legs', 'Len_move': 'std_len'})
walk_feat = pd.concat([d1, d2], axis=1) #Join both dataframes
walk_feat.to_pickle('D:/Baseline/walk.pkl')

#Sitting down
df_sit=df_sit.astype({'Dropoff': 'float64'})
d1=df_sit.groupby(by=["Person", "Irregularity"]).mean() #Mean stored in one dataframe
d1=d1.drop(index=['1','2'], level=1)
d2=df_sit.groupby(by=["Person", "Irregularity"]).std() #Std store in one dataframe
d2=d2.drop(index=['1','2'], level=1)
d2 = d2.rename(columns={'Max_velocity': 'std_velocity', 'Dropoff': 'std_dropoff', 'Len_move': 'std_len'})
sit_feat = pd.concat([d1, d2], axis=1) #Join both dataframes
sit_feat.to_pickle('D:/Baseline/sit.pkl')

#Standing up
df_stand=df_stand.astype({'Dropoff': 'float64'})
d1=df_stand.groupby(by=["Person", "Irregularity"]).mean() #Mean stored in one dataframe
d1=d1.drop(index=['1','2'], level=1)
d2=df_stand.groupby(by=["Person", "Irregularity"]).std() #Std store in one dataframe
d2=d2.drop(index=['1','2'], level=1)
d2 = d2.rename(columns={'Max_velocity': 'std_velocity', 'Dropoff': 'std_dropoff', 'Len_move': 'std_len'})
stand_feat = pd.concat([d1, d2], axis=1) #Join both dataframes
stand_feat.to_pickle('D:/Baseline/stand.pkl')
```
